```python
# You'll need the requests library in order to fully utilize bs4.
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup
import time

df_songs = pd.DataFrame(columns=['movie_title','song_title','lyric_url','song_lyrics'])
```

```python

# Target web page:
url = "https://www.disneyclips.com/lyrics/"
# Establishing the connection to the web page:
response = requests.get(url)
# response.text[:50]
soup = BeautifulSoup(response.text, 'lxml')

# url_1 = "https://www.disneyclips.com/"
```

```python
def song_page (song_url, movie_title):
    # Establishing the connection to the web page:
    time.sleep(1)
    response3 = requests.get(song_url)
    # print(response.text[:50])
    soup3 = BeautifulSoup(response3.text, 'lxml')
    sh1 = soup3.select('h1')[0].get_text(separator="<br>")
    print(sh1)
    sp1 = soup3.select('p')[0].get_text(separator="\n")
    sp1 = sp1.replace("Lyrics transcribed by Disneyclips.com","")
    
    movie_title2 = movie_title.replace(":", " ").replace("/", "")
    sh1 = sh1.replace(":", " ").replace("/", "!")
    
    text_file_name = "songs/" + movie_title2 + "-" + sh1 +".txt"
    
    
#*****************************************************    
    # print(text_file_name)
    # text_file = open(text_file_name, "w")
    # n = text_file.write(sp1)
    # text_file.close()
#*****************************************************       
    
    df_songs.loc[len(df_songs.index)] = [movie_title, sh1, song_url, sp1 ]
    
    
    # print(sp1)
    # sp1.to_csv("songs/test.csv")
    
    
song_page("https://www.disneyclips.com/lyrics/even-when-the-best-part.html","High School Musical: The Musical: The Series Season 2" )    
```

```python
def movie_page (movie_url, movie_title):
    
    
    # Establishing the connection to the web page:
    response2 = requests.get(movie_url,movie_title)
    # print(response.text[:50])
    soup2 = BeautifulSoup(response2.text, 'lxml')
    table = soup2.find('table',{'class':'songs'})
    try:
        for song_link in table.find_all('a'):
            time.sleep(1)
            song_url = song_link.get('href')
            # # song_title = song_link.contents[0]
            # print(f'{song_title}')
            # print(song_link)
            # print(song_page)
            song_url = url + song_url
            print(f'Song Url {song_url}')
            song_page(song_url, movie_title)
         
        
    except:   
        song_page(movie_url,  movie_title)

        
        
    
    
    
    

# movie_page("https://www.disneyclips.com/lyrics/101-dalmatians.html","101 Dalmatians")    
```
