### 08.01.02 Directors & Writers Imports


```python
directors_demo_history = pd.read_csv('../Bens_Data/directors_history.csv')
writers_demo_history = pd.read_csv('../Bens_Data/writers_history.csv')
directors_demo_history.drop(columns=['Unnamed: 0'], inplace=True)
writers_demo_history.drop(columns=['Unnamed: 0'], inplace=True)
```

## 08.02 Merging Data


### 08.02.01 Merging Directors and Principal History


```python
directors_demo_history = pd.merge(directors_demo_history ,                 # left df
                          principals_history[['tconst','primaryTitle', 'runtimeMinutes', 'genres', 'mpaarating', 'rlsdt_dt', 'startYear','rlsdt_mo', 'rlsdt_day', 'rlsdt_daynm','budget_adj','worldwide_adj','ROI', 'averageRating', 'numVotes']],                  # right df
                          how="left",                 # left join
                          left_on='tconst',            # left column
                          right_on='tconst',    # right column
                          indicator = True,           # indicates source of each row
                          #validate = "one_to_many"    # alerts us of the relationship from left to right, incase there are dups
        ) 
```

### 08.02.02 Mergind Writers and Principal History


```python
writers_demo_history = pd.merge(writers_demo_history ,                 # left df
                          principals_history[['tconst','primaryTitle', 'runtimeMinutes', 'genres', 'mpaarating', 'rlsdt_dt', 'startYear','rlsdt_mo', 'rlsdt_day', 'rlsdt_daynm','budget_adj','worldwide_adj','ROI', 'averageRating', 'numVotes']],                  # right df
                          how="left",                 # left join
                          left_on='tconst',            # left column
                          right_on='tconst',    # right column
                          indicator = True,           # indicates source of each row
                          #validate = "one_to_many"    # alerts us of the relationship from left to right, incase there are dups
        ) 
```

## 08.03 Clean Up Merged Data


### 08.03.01 Birth Year and Run Time


```python
directors_demo_history['birthYear'] = pd.to_numeric(directors_demo_history['birthYear'],errors='coerce')
directors_demo_history['runtimeMinutes'] = pd.to_numeric(directors_demo_history['runtimeMinutes'],errors='coerce')
writers_demo_history['birthYear'] = pd.to_numeric(writers_demo_history['birthYear'],errors='coerce')
writers_demo_history['runtimeMinutes'] = pd.to_numeric(writers_demo_history['runtimeMinutes'],errors='coerce')
```
